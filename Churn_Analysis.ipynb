```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df=pd.read_csv('/content/cust churn.csv')
df
```

```python
#converting all the blanks values into 0
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")
df["TotalCharges"]=df["TotalCharges"].astype(float)
```

```python
df.info()
```

```python
#checking for all null values in ds
df.isnull().sum().sum()
```

```python
#for checkong dup values in overall DS.
#REM to check dup values in uniques columns
df.duplicated().sum()
```

```python
#converting 0s/1s into yes/no in senior citizen col
def conv(value):
  if value==1:
    return "yes"
  else:
    return "no"

df['SeniorCitizen']=df['SeniorCitizen'].apply(conv)
```

```python
df.head(10)
```

```python
#plotting bar for churn n count
ax=sns.countplot(x='Churn',data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of Churn")
plt.show()
```

```python
#helps us to understand how many cust hve churned out
gb=df.groupby('Churn').agg({'Churn':"count"})
plt.pie(gb['Churn'],labels=gb.index,autopct="%1.2f%%" )
plt.title("Percentage of Churn")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.title("Churn by Gender")
sns.countplot(x="gender",data=df,hue="Churn")
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title("Churn by SeniorCitizen")
sns.countplot(x="SeniorCitizen",data=df,hue="Churn")
plt.show()
```

```python
churn_counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

churn_counts.plot(kind='bar', stacked=True, figsize=(6,6), colormap='viridis')
plt.title('Stacked Bar Chart of Churn Counts by SeniorCitizen')
plt.xlabel('Senior Citizen')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0) # Rotate x-axis labels to be horizontal
plt.legend(title='Churn')

# Add counts on top of bars
for container in plt.gca().containers:
    plt.bar_label(container)

plt.show()
```
